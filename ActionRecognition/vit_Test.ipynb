```python
import cv2
import os
import matplotlib.pyplot as plt

IMAGE_DIR = "dataset/fanning/train/images"
LABEL_DIR = "dataset/fanning/train/labels"

# IMAGE_DIR = "dataset/trophallaxis/train/images"
# LABEL_DIR = "dataset/trophallaxis/train/labels"
FPS = 60
images = sorted([f for f in os.listdir(IMAGE_DIR) if f.endswith((".jpg", ".png"))])
delay = int(1000 / FPS)
```

```python
def draw_yolo_boxes(img, label_path):
    h, w = img.shape[:2]

    if not os.path.exists(label_path):
        return img

    with open(label_path, "r") as f:
        for line in f:
            cls, xc, yc, bw, bh = map(float, line.split())
            
            # Convert YOLO format to pixel coordinates
            x1 = int((xc - bw / 2) * w)
            y1 = int((yc - bh / 2) * h)
            x2 = int((xc + bw / 2) * w)
            y2 = int((yc + bh / 2) * h)

            # Draw bounding box
            cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)
            
            # Add label text (you'll need to define class names)
            class_name = f"Class {int(cls)}"  # Default label
            # If you have class names, replace the above line with:
            # class_name = CLASS_NAMES[int(cls)]  # where CLASS_NAMES is your list of class names
            
            # Put text on the bounding box
            cv2.putText(img, class_name, (x1, y1 - 10), 
                       cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
    return img
```

```python
for img_name in images:
    img_path = os.path.join(IMAGE_DIR, img_name)
    label_path = os.path.join(LABEL_DIR, img_name.replace(".jpg", ".txt").replace(".png", ".txt"))

    img = cv2.imread(img_path)
    img = draw_yolo_boxes(img, label_path)

    cv2.imshow("Trophollaxis Detection Preview", img)

    key = cv2.waitKey(delay)
    if key == 27:  # ESC to quit
        break

cv2.destroyAllWindows()
```
